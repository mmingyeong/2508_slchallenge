```python
from pathlib import Path
from astropy.io import fits
import os

# 경로 설정
CLASS_PATHS = {
    "slsim_lenses": "/caefs/data/IllustrisTNG/slchallenge/slsim_lenses/slsim_lenses",
    "slsim_nonlenses": "/caefs/data/IllustrisTNG/slchallenge/slsim_nonlenses/slsim_nonlenses",
    "hsc_lenses": "/caefs/data/IllustrisTNG/slchallenge/hsc_lenses/hsc_lenses",
    "hsc_nonlenses": "/caefs/data/IllustrisTNG/slchallenge/hsc_nonlenses/hsc_nonlenses",
}

def human_readable_size(size_bytes):
    units = ["B", "KB", "MB", "GB", "TB"]
    size = float(size_bytes)
    for u in units:
        if size < 1024:
            return f"{size:.2f} {u}"
        size /= 1024
    return f"{size:.2f} PB"

def quick_check(path):
    folder = Path(path)
    files = list(folder.glob("*.fits"))

    # 파일 개수와 총 용량만 계산 (빠름)
    total_files = len(files)
    total_size = sum(f.stat().st_size for f in files[:1000]) * (total_files / min(total_files, 1000))

    # 무결성 샘플 확인 (첫 번째 파일만)
    sample_status = "N/A"
    if files:
        try:
            with fits.open(files[0]) as hdul:
                hdul.verify("exception")
            sample_status = "OK"
        except Exception as e:
            sample_status = f"ERROR: {e}"

    return total_files, human_readable_size(total_size), files[0].name if files else None, sample_status

print("=== 데이터셋 요약 ===")
for cls, path in CLASS_PATHS.items():
    total, size, sample, check = quick_check(path)
    print(f"\n[{cls}]")
    print(f"  경로        : {path}")
    print(f"  파일 개수   : {total}")
    print(f"  추정 총 용량: {size}")
    print(f"  샘플 파일   : {sample}")
    print(f"  무결성 점검 : {check}")
```

```python
from pathlib import Path
from astropy.io import fits
import random

# 클래스별 경로
CLASS_PATHS = {
    "slsim_lenses": "/caefs/data/IllustrisTNG/slchallenge/slsim_lenses/slsim_lenses",
    "slsim_nonlenses": "/caefs/data/IllustrisTNG/slchallenge/slsim_nonlenses/slsim_nonlenses",
    "hsc_lenses": "/caefs/data/IllustrisTNG/slchallenge/hsc_lenses/hsc_lenses",
    "hsc_nonlenses": "/caefs/data/IllustrisTNG/slchallenge/hsc_nonlenses/hsc_nonlenses",
}

def summarize_fits_structure(path):
    folder = Path(path)
    files = list(folder.glob("*.fits"))
    if not files:
        print(f"No FITS files found in {path}")
        return

    sample = random.choice(files)
    print(f"\n=== {sample.name} ===")
    try:
        with fits.open(sample) as hdul:
            # HDU 정보 요약
            hdul.info()

            # 첫 번째 HDU 데이터 구조
            data = hdul[0].data
            if data is not None:
                print(f"Data shape : {data.shape}")
                print(f"Data dtype : {data.dtype}")
                print(f"Min / Max  : {data.min()} / {data.max()}")
            else:
                print("No data in primary HDU")

            # 헤더의 일부 키 샘플
            hdr = hdul[0].header
            print("\nHeader sample:")
            for k in list(hdr.keys())[:10]:
                print(f"  {k:<12}: {hdr[k]}")
    except Exception as e:
        print(f"ERROR reading {sample.name}: {e}")

# 실행
for cls, path in CLASS_PATHS.items():
    print(f"\n######## {cls} ########")
    summarize_fits_structure(path)
```

```python
import numpy as np
from pathlib import Path
from astropy.io import fits
from concurrent.futures import ProcessPoolExecutor
import random
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 경로 설정
# -----------------------------
CLASS_PATHS = {
    "slsim_lenses": "/caefs/data/IllustrisTNG/slchallenge/slsim_lenses/slsim_lenses",
    "slsim_nonlenses": "/caefs/data/IllustrisTNG/slchallenge/slsim_nonlenses/slsim_nonlenses",
    "hsc_lenses": "/caefs/data/IllustrisTNG/slchallenge/hsc_lenses/hsc_lenses",
    "hsc_nonlenses": "/caefs/data/IllustrisTNG/slchallenge/hsc_nonlenses/hsc_nonlenses",
}

SAMPLE_DIR = Path("./pixel_samples")
SAMPLE_DIR.mkdir(exist_ok=True)

# -----------------------------
# 단일 파일 로드 함수
# -----------------------------
def load_single_file(file):
    with fits.open(file) as hdul:
        table = hdul[1].data
        return np.array(table[0][1]).reshape(41, 41).flatten()

# -----------------------------
# 멀티프로세싱 로더
# -----------------------------
def load_pixels_parallel(path, n_samples=5000, n_workers=8):
    folder = Path(path)
    files = sorted(folder.glob("*.fits"))
    selected = random.sample(files, min(n_samples, len(files)))
    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        results = list(executor.map(load_single_file, selected))
    return np.concatenate(results)

# -----------------------------
# 샘플 생성 또는 로드
# -----------------------------
pixel_data = {}
for cls, path in CLASS_PATHS.items():
    save_path = SAMPLE_DIR / f"{cls}_pixels.npy"
    if save_path.exists():
        print(f"Loading cached sample for {cls}...")
        pixels = np.load(save_path)
    else:
        print(f"Creating new sample for {cls}...")
        pixels = load_pixels_parallel(path, n_samples=5000, n_workers=8)
        np.save(save_path, pixels)
    pixel_data[cls] = pixels

# -----------------------------
# 통계 계산
# -----------------------------
stats = []
for cls, data in pixel_data.items():
    stats.append({
        "class": cls,
        "count": len(data),
        "mean": np.mean(data),
        "std": np.std(data),
        "min": np.min(data),
        "max": np.max(data),
        "median": np.median(data),
    })

df_stats = pd.DataFrame(stats)
print("\n=== Pixel Statistics ===")
print(df_stats)

```

```python
# -----------------------------
# 히스토그램 시각화 (x축 제한)
# -----------------------------
plt.figure(figsize=(10, 6))

for cls, data in pixel_data.items():
    plt.hist(data, bins=200, alpha=0.5, label=cls, density=True, histtype="step")

plt.xlabel("Pixel Value")
plt.ylabel("Density")
plt.title("Pixel Value Distribution by Class (5000 samples each)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

# x축 범위 설정: 필요에 따라 조정
#plt.xlim(-10, 200)
plt.yscale("log")

plt.tight_layout()
plt.show()
```
